Q1. What is a problem in computer science?

Answer:
A problem is a task that takes input, processes it using well-defined steps, and produces output.

Q2. What is an algorithm?

Answer:
An algorithm is a finite, ordered set of unambiguous steps to solve a problem.

Q3. What is the difference between a program and an algorithm?

Answer:

Algorithm: Logic / steps (language independent)

Program: Implementation of algorithm in a specific language

Q4. What is a data structure?

Answer:
A data structure is a way of organizing and storing data so it can be accessed and modified efficiently.

Q5. Why do we need data structures?

Answer:
To improve efficiency, reduce time and memory usage, and handle large-scale data.

Q6. Define Primitive data types.

Answer:
Primitive data types store a single value and cannot be divided further (int, float, bool).

Q7. Define Non-Primitive data structures.

Answer:
Non-primitive data structures store multiple values and define relationships between data elements (array, list, dict).

Q8. Is list primitive or non-primitive? Why?

Answer:
Non-primitive, because it stores multiple values and supports operations.

Q9. Is int primitive or non-primitive?

Answer:
Primitive.

Q10. Can an algorithm exist without a data structure?

Answer:
No. Even the simplest algorithm needs some form of data storage.
