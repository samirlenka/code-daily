Q14. Maximum sum subarray of size k.
def max_sum_k(nums, k):
    window = sum(nums[:k])
    max_sum = window
    for i in range(k, len(nums)):

Q15. Longest subarray with sum = k.
def longest_subarray(nums, k):
    prefix = {}
    s = 0
    max_len = 0
    for i, x in enumerate(nums):
        s += x
        if s == k:
            max_len = i + 1
        if s - k in prefix:
            max_len = max(max_len, i - prefix[s-k])
        if s not in prefix:
            prefix[s] = i
    return max_len

    res = [1] * len(nums)
    prefix = 1
    for i in range(len(nums)):
        res[i] = prefix
        prefix *= nums[i]
    suffix = 1

    for x in nums[1:]:
        temp = max_p

Q26. Trapping rain water.
def trap(height):
    l, r = 0, len(height)-1
    left_max = right_max = 0
    water = 0
    while l < r:
        if height[l] < height[r]:
            left_max = max(left_max, height[l])
            water += left_max - height[l]
            l += 1
        else:
            right_max = max(right_max, height[r])
            water += right_max - height[r]
            r -= 1
    return water

Q27. Longest consecutive sequence.
def longest_consecutive(nums):
    s = set(nums)
    longest = 0
    for x in s:
        if x - 1 not in s:
            curr = x
            length = 1
            while curr + 1 in s:
                curr += 1
                length += 1
            longest = max(longest, length)
    return longest

Q31. Find equilibrium index.
def equilibrium(nums):
    total = sum(nums)
    left = 0
    for i, x in enumerate(nums):
        total -= x
        if left == total:
            return i
        left += x
    return -1

Q32. Longest increasing subsequence (n log n).
import bisect
def lis(nums):
    dp = []
    for x in nums:
        pos = bisect.bisect_left(dp, x)
        if pos == len(dp):
            dp.append(x)
        else:
            dp[pos] = x
    return len(dp)

Q33. Rearrange positives and negatives alternately.
def rearrange(nums):
    pos = [x for x in nums if x >= 0]
    neg = [x for x in nums if x < 0]
    res = []
    for i in range(min(len(pos), len(neg))):
        res.extend([pos[i], neg[i]])
    res.extend(pos[i+1:] if len(pos) > len(neg) else neg[i+1:])
    return res

def smallest_subarray(nums, k):
    l = 0
    s = 0
    ans = float('inf')
    for r in range(len(nums)):
        s += nums[r]

def sort_colors(nums):
    low = mid = 0
    high = len(nums) - 1
    while mid <= high:
        if nums[mid] == 0:
            nums[low], nums[mid] = nums[mid], nums[low]
            low += 1
            mid += 1
        elif nums[mid] == 1:
            mid += 1
        else:
            nums[mid], nums[high] = nums[high], nums[mid]
            high -= 1
