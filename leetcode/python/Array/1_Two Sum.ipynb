```python

class Solution:
    def remove_duplicates(self, nums: list[int]) -> int:
        if not nums:
            return 0
        write = 0
        for read in range(1, len(nums)):
            if nums[read] != nums[write]:
                write += 1
                nums[write] = nums[read]
        return write + 1
```

```python
s = Solution()
print(s.remove_duplicates([1,3,4,4,5]))
```
