Solution 1: Brute Force (Naive)
ğŸ’¡ Idea

Try every possible pair and check their sum.

â±ï¸ Complexity

Time: O(nÂ²)

Space: O(1)

âœ… When to use

Very small arrays

Teaching / understanding basics

Worst choice for interviews unless asked first

```python
class Solution:
    def two_sum(self, nums: list[int], target: int)->list[int]:
        n = len(nums)
        print(f"Length of nums: {n}")    
        for i in range(n):
            print(f"i: {i}")
            for j in range(i+1, n):
                print(f"j : {j}")
                if nums[i]+nums[j] == target:
                    return [i, j]
        return "Sum not found"
```

```python
s = Solution()
print(s.two_sum(nums = [1,2,3,4], target = 7))
```

Solution 2: Hash Map (Optimal & Interview Favorite) â­
ğŸ’¡ Idea

For each number:

Calculate its complement

complement = target - nums[i]


Check if complement already exists in a dictionary

If yes â†’ solution found

Otherwise â†’ store current number and index

â±ï¸ Complexity

Time: O(n)

Space: O(n)

âœ… Best choice for interviews
Code (Single Pass Hash Map)

```python
from typing import List

class Solution:
    def twoSum(self, nums: List[int], target: int) -> List[int]:
        seen = {}
        for i, num in enumerate(nums):
            complement = target - num
            if complement in seen:
                return [seen[complement], i]
            seen[num] = i
```

```python
s = Solution()
print(s.twoSum(nums = [1,2,3,4], target = 5))
```

Solution 3: Two-Pointer (Only If Sorted)

âš ï¸ Important: This works only if array is sorted
(LeetCode version is NOT sorted)

ğŸ’¡ Idea

Sort array with original indices

Use two pointers:

left â†’ smallest

right â†’ largest

Adjust pointers based on sum

â±ï¸ Complexity

Time: O(n log n)

Space: O(n)

Code

```python
from typing import List

class Solution:
    def twoSum(self, nums: List[int], target: int) -> List[int]:
        indexed_nums = list(enumerate(nums))
        indexed_nums.sort(key=lambda x: x[1])

        left, right = 0, len(nums) - 1

        while left < right:
            curr_sum = indexed_nums[left][1] + indexed_nums[right][1]
            if curr_sum == target:
                return [indexed_nums[left][0], indexed_nums[right][0]]
            elif curr_sum < target:
                left += 1
            else:
                right -= 1
```

```python
s = Solution()
print(s.twoSum(nums = [6,3,4,4,2], target = 5))
```
